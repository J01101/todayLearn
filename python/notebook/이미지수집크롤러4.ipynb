```python
from bs4 import BeautifulSoup
from selenium import webdriver
import urllib.request
import pandas as pd
import time
import math
import sys
import os
import random

query_txt = input('크롤링 키워드 >> ')
cnt = int(input('크롤링 횟수 입력 >> '))
page_cnt = math.ceil(cnt/10)

f_dir = input('저장 위치 (기본:D:\\SJY\\python\\notebook\\data\\) >> ')
n = time.localtime()
s = '{}-{}-{}-{}-{}-{}'.format(n.tm_year,n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)
result_dir = f_dir+s+'-'+query_txt
os.makedirs(result_dir)

s_time = time.time()
path = "D:/SJY/python/datadown/chromedriver.exe"
```

```python
driver = webdriver.Chrome(path)
driver.get("https://korean.visitkorea.or.kr")

time.sleep(2) # 2초간 페이지를 기다림

try:
    driver.find_element_by_xpath('''//*[@id="chkForm01"]''').click()
except:
    print('코로나 창이 없습니다.')
    
# 검색창의 이름을 찾아서 검색어 입력 후 검색
element = driver.find_element_by_id('inp_search')
element.send_keys(query_txt)
element.send_keys("\n") # 엔터 기능- > 검색 결과페이지로 이동

print("="*80)
print('크롤링할 키워드:',query_txt)
print('크롤링할 총 페이지 번호:',page_cnt)
print("="*80)
```

```python
no = 1            # 게시글 번호 카운트용 변수
no2 =[ ]          # 게시글 번호 저장용 리스트
contents2=[ ]     # 게시글 내용 저장용 리스트

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

for x in range(1,page_cnt+1): # 1페이지부터 반복 시작    
    print("{} 페이지 내용 수집 시작합니다..." .format(x))
    print("="*80)
   
    for i in range(1,11): #페이지당 10개의 페이지를 방문.
        print("="*80)
        print("{} 페이지 {}번째 글 이미지 수집 시작..." .format(x, i))
        print("="*80)
        
        if no > cnt:
            break
            
        try:
            f_name = driver.find_element_by_xpath('''//*[@id="listBody"]/ul/li[{}]/div[2]/div[1]/a'''.format(i)).text
            os.makedirs(result_dir + "\\" + f_name)
            
            driver.find_element_by_xpath('''//*[@id="listBody"]/ul/li[{}]/div[2]/div[1]/a'''.format(i)).click()
        except:
            continue
        
        time.sleep(2)              #페이지가 로딩 될 때까지 2초 대기
    
        html = driver.page_source
        soup = BeautifulSoup(html,'html.parser')
        img_src2 = []
        
        file_no = 0 # 이미지의 순서 + 이미지 이름

        try:
            img_src = soup.find('div','box_txtPhoto').find_all('img')
 
            for i in img_src:
                img_src1 = i['src']
                img_src2.append(img_src1)
        except Exception as e:
            try:
                for y in range(1,5):
                    driver.find_element_by_xpath("""//*[@id="contents"]/div[2]/div[6]/div[1]/div[1]/div/ul/li[{}]/a""".format(y)).click()

                    time.sleep(2)

                    html = driver.page_source
                    soup = BeautifulSoup(html,'html.parser')
                    tag = soup.find('div','cos_cont active')
                    tag2 = tag.find('div','info_area pc').find_all('a')

                    for i in tag2:
                        style = i['style'].split('(')[1].split(')')[0] #style속성안에 괄호안에있는거
                        img_src2.append(style)

            except Exception as e:
                print(e);print('유형3')
        
        #이미지 다운로드 받을 폴더로 이동
        os.chdir(result_dir + '\\' +  f_name)

        #이미지 다운로드 실행
        for i in range(0, len(img_src2)):
            print('{}번째 이미지 저장 중...'.format(file_no))
            try:
                urllib.request.urlretrieve(img_src2[i], str(file_no) + '.jpg')
            except:
                continue
            file_no += 1
            time.sleep(0.5)
    
        driver.back()
        no += 1
    
    if x > page_cnt + 1:
        break
        
    x += 1
              
    time.sleep(1)   

    # 페이지 넘기기
    if(x % 5 == 1):
        driver.find_element_by_link_text("다음").click()
    else:
        driver.find_element_by_link_text("{}".format(x)).click()
        
    time.sleep(2)
    
    
```
