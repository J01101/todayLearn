```python
f = open('./files/stockcode.txt', 'r')
line_num = 1
line = f.readline() #텍스트파일에서 한줄을 읽는다.
while line: #line의 값이 빈 문자열일 때까지 반복한다.
   print('%d %s' %(line_num, line), end='')
   line = f.readline() #텍스트파일에서 한줄을 읽는다.
   line_num += 1 #라인넘버에 1을 더한다.
f.close()
```

```python
num = input('아무 숫자를 입력하세요: ')
 
if num.isdigit():#입력한 값이 숫자로 구성되었다면
   num = num[::-1] #뒤에서부터 거꾸로 가져온다.00015
   ret = ''
   for i, c in enumerate(num): #인덱스와 값으로 추출
      i += 1
      if i != len(num) and i%3 == 0: #요소의 위치가 num의 마지막이 아니고 3의 배수이면 
         ret += (c + ',') #추출한 요소 뒤 콤마를 연결한 값을 ret에 추가한다.
      else:
         ret += c
   ret = ret[::-1] #ret값을 거꾸로 가져온다. 15,000
   print(ret)
else:
   print('입력한 내용 [%s]: 숫자가 아닙니다.' %num)
```

```python
#스택 구현하기(append, pop) 스택은 차곡차곡 쌓이듯이 올라가서 나중에 저장된 자료가 먼저 추출될 수 있도록 되어 있는 나열구조이다.

mystack = [] #스택으로 사용될 리스트
 
def putdata(data): #data를 전역변수 리스트에 저장하는 함수
   global mystack
   mystack.append(data)
```

```python
def popdata(): #데이터가 있으면 
   global mystack
   if len(mystack) == 0:
      return None
   return mystack.pop() #pop()은 리스트의 맨 마지막 요소를 돌려주고 그 요소는 삭제한다.
```

```python
putdata('데이터1')
putdata([3, 4, 5, 6])
putdata(12345)
```

```python
print('<스택상태>: ', end=''); print(mystack)
 
ret = popdata()
while ret != None:    
    print('스택에서 데이터 추출:', end=''); print(ret)
    print('<스택상태>: ', end=''); print(mystack)
    ret = popdata()
```

```python
# 문장 문자 빈도수 계산 - 파일이름을 인자로 받아 이 파일에 있는 내용에서 각 문자들이 나타나는 빈도수를 문자별로 계산하여 그 결과를 사전으로 리턴한다.
def getTextFreq(filename):
   with open(filename, 'r') as f:
      text = f.read()
      fa = {} #빈사전
      for c in text: #문자 하나씩 추출
         if c in fa: #추출한 문자가 fa사전에 있는지 확인
            fa[c] += 1 #있으면 문자키값에 1을 증가
         else:
            fa[c] = 1 #첫문자이면 그문자키값에 1을 할당한다.
   return fa
```

```python
ret = getTextFreq('./files/stockcode.txt')
ret = sorted(ret.items(), key=lambda x:x[1], reverse=True) #빈도값을 기준으로 내림차순으로 정렬한다.
for c, freq in ret:
   if c == '\n': #줄바꿈 기호는 넘어간다.(무시한다.)
      continue
   print('[%c] -> [%d]회 나타남' %(c, freq))
```

```python
def countWord(filename, word):
   with open(filename, encoding='utf-8') as f:
      text = f.read()
      text = text.lower() #소문자로 변경
      pos = text.find(word) #word의 값이 최초로 나타나는 위치를 가져온다.
      count = 0
      while pos != -1: #word의 위치가 있는 경우
         count += 1 #count에 1을 증가하고
         pos = text.find(word, pos+1) #두번째 인자부터 다시 word의 값 검색
   return count
 
word = input('mydata.txt에서 개수를 구할 단어를 입력하세요: ')
word = word.lower()
ret = countWord('./files/mydata.txt', word)
print('[%s]의 개수: %d' %(word, ret))
```

```python
# URL에 접속하여 HTML 페이지를 화면에 출력하기, 파일로 저장하기 urllib.request 모듈의 urlopen(url)은 url에 접속하여 파일처럼 다룰 수 다.
from urllib.request import urlopen
 
url = 'https://www.python.org'
with urlopen(url) as f:
    doc = f.read().decode() #read()는 바이트스트림으로 읽어온다. decode는 텍스트로 변환하여준다.
    print(doc)
```

```python
'''
- 인터넷에 있는 대용량 파일을 내 PC로 저장하기
  영화 파일과 같이 1GB가 넘는 대용량 파일을 한 번에 읽어 내 PC로 
  저장하려면 메로리 문제가 발생할 수 있다. 
  대용량 파일을 250KB씩 읽어 내 PC로 저장한다.
'''

from urllib.request import urlopen
 
BUFSIZE = 256*1024
 
fileurl = 'https://www.python.org/ftp/python/3.5.2/python-3.5.2.exe'
filename = fileurl.split('/')[-1]
try:
    with urlopen(fileurl) as f:
        with open("../files/"+filename, 'wb') as h:
            buf = f.read(BUFSIZE)
            while buf:
                h.write(buf)
                buf = f.read(BUFSIZE)                
except Exception as e:
    print(e)
```
