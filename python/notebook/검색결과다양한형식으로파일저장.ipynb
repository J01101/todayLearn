```python
 # Step 1. 필요한 모듈과 라이브러리를 로딩하고 검색어를 입력 받습니다
from bs4 import BeautifulSoup
from selenium import webdriver
import time
import sys
# 분류 저장을 위한 import
import xlwt
import pandas as pd

query_txt = input('크롤링 키워드 >> ')
ft_name = input('검색 결과를 저장할 txt 파일경로와 이름\n(Ex_D:\\SJY\\python\\notebook\\data\\test.txt) >> ')
fc_name = input('검색 결과를 저장할 csv 파일경로와 이름\n(Ex_D:\\SJY\\python\\notebook\\data\\test.csv) >> ')
fx_name = input('검색 결과를 저장할 xlsx 파일경로와 이름\n(Ex_D:\\SJY\\python\\notebook\\data\\test.xlsx) >> ')

path = "D:/SJY/python/datadown/chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://korean.visitkorea.or.kr")

time.sleep(2) # 2초간 페이지를 기다림

try:
    driver.find_element_by_xpath('''//*[@id="chkForm01"]''').click()
except:
    print('코로나 창이 없습니다.')
    
# 검색창의 이름을 찾아서 검색어 입력 후 검색
element = driver.find_element_by_id('inp_search')
element.send_keys(query_txt)
element.send_keys("\n") # 엔터 기능- > 검색 결과페이지로 이동

time.sleep(1) # 1초간 페이지를 기다림 (결과 페이지를 기다림.)

# 페이지 내에 있는 내용을 화면에 출력
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

content_list = soup.find('ul', 'list_thumType type1')

# Step 5. 각 항목별로 분리하여 추출하고 변수에 할당하기
no = 1
no2 =[ ]
contents2=[ ]
tags2=[ ]

# 분류해서 받아오기 + 출력
for i in content_list:
    no2.append(no)
    print('번호', no)
    
    if i.find('div', 'tit'):
        contents = i.find('div', 'tit').get_text()
        contents2.append(contents) # 표로 만들어서 출력하기 위해
        print('내용', contents.strip())
        
    else:
        contents = '내용 없음'
        contents2.append(contents) # 표로 만들어서 출력하기 위해
        print('내용', contents.strip())
        
    if i.find('p', 'tag_type'):
        tags = i.find('p', 'tag_type').get_text()
        tags2.append(tags) # 표로 만들어서 출력하기 위해
        print('태그', tags.strip())
    else:
        tags = '내용 없음'
        tags2.append(tags) # 표로 만들어서 출력하기 위해
        print('태그', tags.strip())
        
    print("\n")
    no += 1
    
# 학습목표 2: 분리 수집된 데이터를 데이터 프레임으로 만들어서 
# csv , xls 형식으로 저장하기  
# 출력 결과를 표(데이터 프레임) 형태로 만들기
korea_trip = pd.DataFrame()
korea_trip['번호']=no2
korea_trip['내용']=contents2
korea_trip['태그']=tags2

korea_trip.to_csv(fc_name, encoding='utf-8-sig') # sig를 안쓰면 엑셀로 깨짐
print(" csv 파일 저장 경로: {}" .format(fc_name))

# openpyxl로 사용을 권장하기때문에 에러가 발생함 이게 싫으면 아래의 코드의 주석을 풀면됨.
# import warnings
# warnings.filterwarnings(action='ignore'
korea_trip.to_excel(fx_name)
print(" xls 파일 저장 경로: {}" .format(fx_name))

# 출력 결과를 txt 파일로 저장하기
f = open(ft_name, 'a',encoding='UTF-8')
f.write(str(contents2))
f.write(str(tags2))
f.close( )
print(" txt 파일 저장 경로: %s" %ft_name)  

#driver.close()
```
