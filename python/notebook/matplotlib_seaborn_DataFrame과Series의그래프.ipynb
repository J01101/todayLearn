### matplotlib로 그래프 그리기

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips = tips.rename({'sex':'gender'},axis='columns')
tips.head()
```

### 다변량 그래프 - 3개 이상의 변수를 사용

```python
def recode_gender(gender):
    if gender == 'Female':
        return 0 
    else:
        return 1
```

```python
tips['gender_color'] = tips['gender'].apply(recode_gender)
 
tips.head()
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size']*10,
    c = tips['gender_color'],
    alpha = 0.5
)
axes1.set_title("Total Bill vs Tip Colored by gender and Sized by size")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
```

### seaborn 라이브러리 활용하기

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips = tips.rename({'sex':'gender'},axis='columns')
tips.head()
```

```python
import warnings

warnings.filterwarnings(action='ignore')
```

```python
ax = plt.subplot()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
ax = plt.subplot()
ax = sns.distplot(tips['total_bill'], kde=False)
ax.set_title('Total Bill Histogram')
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Frequency")
```

```python
ax = plt.subplot()
ax = sns.distplot(tips['total_bill'], hist=False)
ax.set_title('Total Bill Density')
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Unit Probability")
```

```python
hist_den_rug, ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
ax.set_title('Total Bill Histogram with Density and Rug Plot')
ax.set_xlabel('Total Bill')
```

```python
x = plt.subplots()
ax = sns.countplot('day', data=tips)
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

### 다양한 종류의 이변량 그래프 그리기

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data=tips)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data=tips, fit_reg = False)
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
joint = sns.jointplot(x='total_bill', y='tip', data=tips)
joint.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total Bill and Tip', fontsize=13, y=1.03)
```

```python
hexbin = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
hexbin.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
hexbin.fig.suptitle('Hexbin of Total Bill and Tip', fontsize=10, y=1.03)
```

```python
ax = plt.subplot()
ax = sns.kdeplot(
    data = tips['total_bill'],
    data2 = tips['tip'],
    shade=True
)
ax.set_title('Kernel Density Plot of Total Bill and Tip')
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
```

```python
ax = plt.subplot()
ax = sns.barplot(
    x = 'time',
    y = 'total_bill',
    data=tips
)
ax.set_title('Bar Plot of average total bill for time of day')
axes1.set_xlabel("Time of day")
axes1.set_ylabel("Average total bill")
```

```python
ax = plt.subplot()
ax = sns.boxplot(
    x = 'time',
    y = 'total_bill',
    data=tips
)
ax.set_title('Boxplot of total bill by time of day')
axes1.set_xlabel("Time of day")
axes1.set_ylabel("Total Bill")
```

```python
ax = plt.subplot()
ax = sns.violinplot(
    x = 'time',
    y = 'total_bill',
    data=tips
)
ax.set_title('Violinplot of total bill by time of day')
axes1.set_xlabel("Time of day")
axes1.set_ylabel("Total Bill")
```

```python
sns.pairplot(tips)
```

```python
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
```

### 다변량 그래프

```python
sns.lmplot(x='total_bill', y='tip', data=tips, 
           fit_reg=False, col='day', col_wrap=2) #col_wrap 한행에 2개씩그린다.
plt.show()
```

```python
facet = sns.FacetGrid(tips, col='time')
facet.map(sns.distplot, 'total_bill', rug=True)
plt.show()
```

```python
facet = sns.FacetGrid(tips, col='day', hue='gender', col_wrap=2)
facet.map(plt.scatter, 'total_bill', 'tip')
facet.add_legend()
plt.show()
```

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='gender')
facet.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```
