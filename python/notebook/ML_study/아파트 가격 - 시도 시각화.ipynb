```python
import pandas as pd
import os
```

* 2020년 01월 국토부 매매데이터를 읽어온다.

```python
total = []
for roots, dirs, files in os.walk("./apt"):
    for file in files:
        total.append(pd.read_csv(roots + "/" + file,
                  header=[15], encoding='euc-kr', low_memory=False))
```

```python
df = pd.concat(total)
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df['거래금액(만원)'] = df['거래금액(만원)'].apply(lambda x : int(x.replace(",", "")))
```

```python
df.sort_values(by = ['거래금액(만원)'], ascending=False, inplace=True)
```

```python
df
```

* 서울의 데이터만 따로 뽑기 위해서 시군구 컬럼에 '서울'이 포함된 row를 선택한다.

```python
seoul_df = df[df.시군구.str.contains('서울')].copy()
```

```python
seoul_df.shape
```

* 서울 데이터에서 구별로 평균을 계산하기 위해서 '구' 컬럼을 따로 생성 

```python
seoul_df
```

```python
seoul_df['구'] = seoul_df['시군구'].apply(lambda x : x.split()[1])
```

```python
seoul_df.head(5)
```

* groupby를 계산하고 대상이 index로 이동하는 것을 막기 위해서 as_index 옵션을 False로 설정
* as_index 값은 기본값으로 True

```python
seoul_gu = seoul_df.groupby(['구'], as_index=False)['거래금액(만원)'].agg(['mean',
                                                    'max', 'min', 'std'])
```

```python
seoul_gu.reset_index(inplace=True)
```

```python
seoul_gu
```

* 지도를 그리기 위한 데이터를 불러 들인다. skorea_municipal... 이 데이터는 서울시 행정구역 위치에 설정한 파일이며 
* 파일의 출처는 https://github.com/southkorea/southkorea-maps/blob/master/kostat/2013/json/skorea_municipalities_geo_simple.json
* 이 데이터를 읽어서 처리하기 위해서는 folium이라는 패키지가 필요하니 만약 없으면 설치해야 한다. 

```python
import json
import folium
import warnings
warnings.simplefilter(action = "ignore", category = FutureWarning)
geo_path = './skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))
```

```python
geo_str
```

* 지도위에 서울를 설정한다.

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')
```

```python
map
```

* 지도 위에 choropleth 함수에 거래금액을 넘겨 지도위에 색깔을 입힌다. 

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11, tiles='Stamen Toner')
map.choropleth(geo_data = geo_str,
               data = seoul_gu,
               columns = ['구', 'mean'],
               fill_color = 'PuRd', #PuRd, YlGnBu
               key_on = 'feature.id')
folium.LayerControl().add_to(map)
```

* 해당 결과를 아래 html 파일 형식으로 저장한다. 

```python
map.save("./2020_01_map_test.html")
```
