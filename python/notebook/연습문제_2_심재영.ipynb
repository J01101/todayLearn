```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
import math
import sys
```

```python
keyword = input('키워드 >> ')

save_path = 'D:\\SJY\\python\\notebook\\crawlingData\\'
ft_name = input('텍스트 파일 이름 >> ')
ft_name = save_path + ft_name + '.txt'
fc_name = input('csv 파일 이름 >> ')
fc_name = save_path + fc_name + '.csv'
fx_name = input('xlsx 파일 이름 >> ')
fx_name = save_path + fx_name + '.xlsx'
```

```python
path = "D:\\SJY\\python\\datadown\\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://naver.com")
time.sleep(2) # 2초간 페이지를 기다림
```

```python
# 검색창의 이름을 찾아서 검색어 입력 후 검색
element = driver.find_element_by_xpath('//*[@id="query"]')
element.send_keys(keyword)
element.send_keys("\n")
```

```python
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
no = 1
list_no=[]
list_title=[]
list_content=[]
list_wdate=[]
list_nname=[]

list_total = soup.find_all('li', 'bx','_svp_item')

for i in list_total:
    list_no.append(no)
    print('번호 / ', no)
    
    if i.find('a', 'api_txt_lines', 'total_tit'):
        title = i.find('a', 'api_txt_lines', 'total_tit').get_text()
        list_title.append(title)
        print('제목 / ', title)
    else:
        title = '정보없음'
        list_title.append(title)
    
    if i.find('div', 'api_txt_lines', 'dsc_txt'):
        content = i.find('div', 'api_txt_lines', 'dsc_txt').get_text()
        list_content.append(content)
        print('내용 / ', content)
    else:
        content = '정보없음'
        list_content.append(content)
        
    if i.find('span', 'sub_time', 'sub_text'):
        wdate = i.find('span', 'sub_time', 'sub_text').get_text()
        list_wdate.append(wdate)
        print('작성일 / ', wdate)
    else:
        wdate = '정보없음'
        list_wdate.append(wdate)
    
    if i.find('a', 'sub_txt', 'sub_name'):
        nname = i.find('a', 'sub_txt', 'sub_name').get_text()
        list_nname.append(nname)
        print('작성자 / ', nname)
    else:
        nname = '정보없음'
        list_nname.append(nname)
    
    print('\n')    
    no+=1
```

```python
trip = pd.DataFrame()
trip['번호']=list_no
trip['제목']=list_title
trip['내용']=list_content
trip['작성일']=list_wdate
trip['작성자']=list_nname

trip.to_csv(fc_name, encoding='utf-8-sig')
print(" csv 파일 저장 경로: {}" .format(fc_name))

trip.to_excel(fx_name)
print(" xlsx 파일 저장 경로: {}" .format(fx_name))

f = open(ft_name, 'a',encoding='UTF-8')
f.write('번호 / '+str(list_no)+'\n')
f.write('제목 / '+str(list_title)+'\n')
f.write('내용 / '+str(list_content)+'\n')
f.write('작성일 / '+str(list_wdate)+'\n')
f.write('작성자 / '+str(list_nname)+'\n')
f.close( )
print(" txt 파일 저장 경로: %s" %ft_name)  
```
