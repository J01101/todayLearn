### concat 메서드로 데이터 연결하기

```python
import pandas as pd
 
df1 = pd.read_csv('data2/concat_1.csv') 
df2 = pd.read_csv('data2/concat_2.csv') 
df3 = pd.read_csv('data2/concat_3.csv')
 
df1,df2,df3
```

```python
row_concat = pd.concat([df1, df2, df3]) #데이터프레임 아래로 연결
row_concat
```

```python
row_concat.iloc[3, ] #3번째 행의 모든 열의 값 추출
```

```python
new_row_series = pd.Series(['n1', 'n2', 'n3', 'n4'])
pd.concat([df1, new_row_series])
```

```python
new_row_df = pd.DataFrame([['n1', 'n2', 'n3', 'n4']], columns=['A', 'B', 'C', 'D']) 
new_row_df
```

```python
pd.concat([df1, new_row_df])
```

```python
df1.append(new_row_df)
```

```python
# ignore_index = True로 지정하면 데이터 통합후 인덱스를 0부터 다시 지정한다.
data_dict = {'A': 'n1', 'B': 'n2', 'C': 'n3', 'D': 'n4'}
print(df1.append(data_dict, ignore_index=True))
```

### 열 방향으로 데이터 연결하기 

```python
col_concat = pd.concat([df1,df2,df3], axis = 1)
col_concat
```

```python
col_concat['A']
```

```python
col_concat['new_col_list'] = ['n1','n2','n3','n4']
col_concat
```

```python
#데이터프레임 열 방향으로 연결하여 열 이름이 중복될때 ingnore_index = True로 지정하면 열이름을 0부터 다시 지정된다.

pd.concat([df1, df2, df3], axis=1, ignore_index=True)
```

### merge로 옆으로(열방향) 통합하기

```python
person = pd.read_csv('data2/survey_person.csv')
site = pd.read_csv('data2/survey_site.csv')
survey = pd.read_csv('data2/survey_survey.csv')
visited = pd.read_csv('data2/survey_visited.csv')
```

```python
person
```

```python
site
```

```python
visited
```

```python
survey
```

```python
visited_subset = visited.loc[[0,2,6], ] # 0,2,6인덱스
visited_subset
```

```python
o2o_merge = site.merge(visited_subset, left_on='name', right_on='site')
o2o_merge
```

```python
m2o_merge = site.merge(visited, left_on='name', right_on='site')
m2o_merge
```

```python
ps = person.merge(survey, left_on='ident', right_on='person')
ps
```

```python
vs = visited.merge(survey, left_on='ident', right_on='taken')
vs
```

```python
ps_vs = ps.merge(vs, left_on=['ident','taken','quant','reading'],
                right_on=['person','ident','quant','reading'])
ps_vs.loc[0,]
```

```python
ps_vs.head()
```

```python
sample_1 = pd.read_excel('data2/sample_1.xlsx', 
                         header=1,
                         skipfooter=2, 
                         usecols='A:C')
sample_1.head(3)
```

```python
code_master = pd.read_excel('data2/sample_codemaster.xlsx')
code_master
```

```python
sample_1_code = pd.merge(left=sample_1,
                        right=code_master,
                        how = 'left',
                        left_on='국적코드',
                        right_on='국적코드')
sample_1_code
```

```python
sample_1_code = pd.merge(left=sample_1,
                        right=code_master,
                        how = 'inner',
                        left_on='국적코드',
                        right_on='국적코드')
sample_1_code
```

### append 메서드로 아래로(행방향)통합하기

```python
sample_1_code['기준년월'] = '2019-11'
sample_1_code
```

```python
sample_2 = pd.read_excel('data2/sample_2.xlsx',
                        header=1,
                        skipfooter=2,
                        usecols='A:C')

sample_2['기준년월'] = '2019-12'
```

```python
sample_2_code = pd.merge(left=sample_2,
                        right=code_master,
                        how = 'left',
                        left_on='국적코드',
                        right_on='국적코드')
sample_2_code
```

```python
sample = sample_1_code.append(sample_2_code, ignore_index=True)
sample
```

```python
# sample 데이터 프레임 엑셀 파일로 저장
# !pip install openpyxl  
sample.to_excel('data2/sample.xlsx', index=False) #인덱스번호 제외하고 저장
```
