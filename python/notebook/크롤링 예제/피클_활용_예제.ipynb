환율은 1121원

```python
kbo = test

import pandas as pd
pd.set_option('display.float_format', lambda x : '%.3f' % x)
```

```python
# 1. 고양 히어로즈 -> 키움 히어로즈 이름 변경
kbo.loc[kbo['Team'] == '고양 히어로즈', 'Team'] = '키움 히어로즈'
```

```python
# 2. 신장/체중 -> 신장과 체중 컬럼 생성 (단위생략)
kbo['신장'] = kbo['신장/체중'].apply(lambda x : x.split('/')[0][:-2])
```

```python
kbo['체중'] = kbo['신장/체중'].apply(lambda x : x.split('/')[1][:-2])
```

```python
kbo
```

```python
# drop 신장/체중
kbo.drop('신장/체중', axis=1, inplace = True)
```

```python
kbo
```

```python
#3. BMI 컬럼 생성
kbo['신장'] = kbo['신장'].astype(int)
```

```python
kbo['체중'] = kbo['체중'].astype(int)
```

```python
kbo['BMI'] = kbo['체중']/ (kbo['신장']/100) ** 2
```

```python
kbo
```

```python
kbo.sort_values(by=['BMI'], ascending=False).head(5).reset_index(drop=True)
```

```python
# 4. 나이 컬럼 생성
kbo['나이'] = kbo['생년월일'].apply(lambda x : 2021- int(x[:4]) + 1)
```

```python
kbo.sort_values(by=['나이'], ascending=False).head(5)
```

```python
kbo.groupby('Team')[['나이']].mean()
```

```python
# 5. 경력 컬럼 생성
kbo['입단년도'].apply(lambda x : int(x[:2]) + 1)
```

```python
# 입단년도가 없는 사람들 수 체크
kbo.loc[kbo['입단년도'] == '', '입단년도'] = kbo.loc[kbo['입단년도'] == '', '지명순위'].apply(lambda x : x[:2])
```

```python
kbo.loc[kbo['입단년도'] == '']
```

```python
def carr(x):
    try:
        x = int(x[:2])
        if x > 82:
            return 2021 - (x + 1900) + 1
        else:
            return 21-x + 1
    except:
        return x
```

```python
kbo['경력'] = kbo['입단년도'].apply(lambda x : carr(x))
```

```python
kbo[kbo['선수명'] == '송승준']
```

```python
kbo
```

```python
def salary(x):
    if x.find('달러') > -1:
        return int(x[:-2]) * 1121
    elif x.find('만원') > -1:
        return int(x[:-2]) * 10000
```

```python
kbo['연봉2'] = kbo['연봉'].apply(lambda x : salary(x))
```

```python
kbo
```

```python
# 6. 구단별 연봉의 최고, 평균, 중앙값 출력
kbo.groupby(['Team'])['연봉2'].agg(['max','mean','median','std'])
```

```python
# 7. 포지션별 연봉 평균
kbo.groupby('포지션')['연봉2'].mean().sort_values()
```
